```python
import pandas as pd
from pprint import pprint
```

```python
df = pd.read_excel(
    'Переводческий_3_2019_2020_303_2_ан-нем_1_сентября_-_8_сентября.xls',
    skiprows=[0], 
    names=['Day', 'Time', 'Discipline', 'Audience'])
df['Day'] = pd.Series(df['Day']).fillna(method='ffill')
```

```python
df
```

```python
df.groupby(['Day']).get_group('ПН').loc[0]
```

```python
df_by_day = df.groupby(['Day'])
```

```python
schedule = {}

for group in df_by_day.groups:
    day = df.groupby(['Day']).get_group(group)
    
    schedule[group] = []
    
    for index, lesson in day.iterrows():
        teacher = None
        subject = None
        
        if str(lesson.Discipline) != 'nan':
            discipline = str(lesson.Discipline).split('\n')
            subject = discipline[0]
            if len(discipline) > 1:
                teacher = str(lesson.Discipline).split('\n')[1]
        
        schedule[group].append({
            #'Day': lesson.Day,
            'Time': lesson.Time,
            'Audience': lesson.Audience,
            'Teacher': teacher,
            'Subject': subject
        })
    
    schedule[group].sort(key=lambda x: x['Time'])


pprint(schedule)
```
